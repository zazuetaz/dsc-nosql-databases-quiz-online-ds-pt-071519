## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

SQL is relational db - we expect data to have standard format and it can be split across multiple tables that can be tied together with a query.

NoSQL is a non-relational db - data is stored in different structures. You might use a NoSQL if you don't know what the size or shape of your data might be.  You would also use a NoSQL db if you were operating on Big Data and wanted to significantly reduce processing speed for queries.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB is a Document store, which means that data is stored in documents and then within each document you can store metadata. The data is structured in key-value pairs that allow you to query the metadata by calling the key. This format allows us to store data quickly, even if we don't know how we want to deal with it or what it looks like, by assigning it a key.

If we tried to do this in an rdb, we would need to better understand the structure of the data and what it meant to us/how we would want to use it before assigning it to a table column.

An example could be your full history of text message conversations on your phone. Conversations vary in frequency, duration, and length. The messages you send will contain different information as well. You could put these in a Document Store; each contact could be a documet; each date you had correspondence could be an additional document within the contact document; each message within the daily correspondence could be a sub-document within the daily document; and you could further break down the data from there.




## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

If you are pulling data from a number of servers, and a few go down, it could be difficult for you to know that you were not pulling fresh data, or you might not be able to run your queries at all until the servers were restored. RDDs take parts of the data on each server and store backups on the other servers. Then each server is responsible for running a subset of the data, but can be tapped into for a missing piece if the server responsible for running that piece goes down. This allows your system to have Fault Tolerance - you can still be operational if some of your servers are down.




## Question 5

What is MapReduce? How is it related to Hadoop?

_______________________________________________________________________________________________________________________________
Hadoop is an instance of an RDD. What it does is ask many small servers that are responsible for storing similar information to trigger the same query at the same time. That is the Map portion. Reduce then rolls up the identical queries into one output, thereby allowing you to reduce the query time for a massive amount of information.
